```python
from prophet import Prophet

# Prepare time series
trend_ds = ds.groupby('Date of Report').size().reset_index()
trend_ds.columns = ['ds', 'y']

model = Prophet()
model.fit(trend_ds)

future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

model.plot(forecast)
plt.title("Forecasted Child Abuse Reports")
plt.show()
```

```python
ds=ds[['District', 'Type of Abuse',
       'Reporter Type', 'Emotional', 'Neglect', 'Physical', 'Sexual', 'Risk_Level',
       'all abuse', 'abuse_type']]
```

```python
ds.columns
```

```python
x_ds = ds[['District', 'Emotional', 'Neglect', 'Physical','Sexual' , 'Reporter Type', 'Risk_Level']]

y_ds = ds[['Type of Abuse']]
```

```python
ds.dtypes
```

```python
x_ds
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_ds, y_ds, test_size=0.2, random_state=42, stratify=y_ds)
print(f"Training set size: {x_train.shape}, Testing set size: {x_test.shape}")
```

```python
y_train.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(max_iter=1000).fit(x_train, y_train.values.ravel())
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42).fit(x_train, y_train.values.ravel())
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = lr_model.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_predict = rf_model.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_predict))
print(classification_report(y_test, y_predict))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(rf_model, x_ds, y_ds.values.ravel(), cv=5)

print("Cross-validation scores:", scores)
print("Average CV accuracy:", scores.mean())
```
